```python
rom IPython import get_ipython
from IPython.display import display
# %%
import requests
import pandas as pd
import time
import base64
from collections import Counter, defaultdict

# GitHub headers with personal access token
headers = {
    'Accept': 'application/vnd.github.v3+json',
    'Authorization': 'Bearer '
}

ai_keywords = [
    # Core AI and Copilot-like coding terms
    'copilot', 'AI-generated', 'ChatGPT', 'automated code', 'AI assistant',
    'artificial intelligence', 'machine learning', 'deep learning', 'neural network',
    'natural language processing', 'autonomous', 'robotics', 'prompt engineering',
    'generative AI', 'LLM', 'language model', 'code suggestion', 'code completion',
    'smart autocomplete', 'AI coding', 'AI-powered', 'AI developer tool',
    'intelligent code generation', 'auto code review', 'auto refactoring',
    'AI-driven', 'pair programmer', 'code copilot', 'AI-enhanced coding',
    'prompt-based development', 'AI-based recommendation', 'AI programming assistant',
    'virtual programmer', 'synthetic code', 'AI model', 'program synthesis',
    'auto code generation', 'AI pair programming', 'voice coding', 'natural code input',
    'code automation', 'automated coding', 'AI-supported coding', 'AI software engineering',

    # Well-known tools and platforms
    'GitHub Copilot', 'Amazon CodeWhisperer', 'Tabnine', 'CodeGeeX',
    'Codota', 'Sourcery', 'PolyCoder', 'OpenAI Codex',
    'Cursor AI', 'Codeium', 'Refact.ai', 'Continue', 'Kite',
    'AI21 Studio', 'Replit Ghostwriter', 'DeepCode', 'Mutable AI',
    'JetBrains AI Assistant', 'AWS CodeWhisperer', 'Phind'
]
contribution_files = ['CONTRIBUTING.md', 'docs/CONTRIBUTING.md', '.github/CONTRIBUTING.md'] # Define the list of contribution files

# Your list of organizations
orgs = [ "gatsbyjs", "commaai", "minio", "pmndrs",
"ionic-team", "google", "meilisearch", "Semantic-Org", "mozilla", "JetBrains", "facebookresearch",
"appwrite", "PaddlePaddle", "TryGhost", "ethereum", "cypress-io", "PowerShell", "serverless",
"square", "x64dbg", "hashicorp", "android", "meteor", "ColorlibHQ", "discourse", "RocketChat",
"prisma", "metabase", "THUDM", "Stability-AI", "getsentry", "Kong", "apachecn"

]

def check_contribution_guidelines(full_name):
    for file_name in contribution_files:
        file_url = f'https://api.github.com/repos/{full_name}/contents/{file_name}'
        file_resp = requests.get(file_url, headers=headers)
        if file_resp.status_code == 200:
            print(f'üìÑ Checking file: {file_url}')
            content = file_resp.json().get('content', '')
            if content:
                decoded = base64.b64decode(content).decode('utf-8', errors='ignore')
                matching_keywords = [kw for kw in ai_keywords if kw.lower() in decoded.lower()]
                if matching_keywords:
                    print(f'\nüîç Repository: {full_name}')
                    print(f'üìÑ File: {file_url}')
                    print('‚ö†Ô∏è  Found AI-related mention!\n')
                    print(f'Matching words: {", ".join(matching_keywords)}\n')
                    break
        elif file_resp.status_code != 404:
            print(f'‚ùå Failed to fetch {file_url} (Status: {file_resp.status_code})')
for org in orgs:
    url = f"https://api.github.com/orgs/{org}/repos?per_page=100&page=1"
    res = requests.get(url, headers=headers)
    if res.status_code != 200 or not res.json():
        print(f'‚ùå Failed to fetch {url} (Status: {res.status_code})')
        continue

    for repo in res.json():
        repo_name = repo["name"]
        check_contribution_guidelines(org + '/' + repo_name)
        time.sleep(0.1)  # Add a small delay

    time.sleep(0.2)  # Add a small delay between organizations
```
